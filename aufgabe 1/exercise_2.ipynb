```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
""" 2 a)"""
data = np.load('upb.npy')
#print(data.shape) It prints (770, 1894, 3) height * width * depth/channels (3 for RGB)
```

```python
""" 2 b)""" 
plt.imshow(data)
plt.axis('off')
plt.show()      # shows image of UPB logo
```

```python
""" 2 c)"""      
# shows total number of elements in the array which is 770 * 1894 * 3
print(data.size)
```

```python
""" 2 d)"""
# Show red, green, blue channels separately (vertically stacked):

fig, axs = plt.subplots(3, 1, figsize=(5, 15))

axs[0].imshow(data[:, :, 0], cmap='Reds') #take all rows, all cols, but onyl 0rd channel which is read + use red color map for visualizing
axs[0].set_title('Red Channel')

axs[1].imshow(data[:, :, 1], cmap='Greens')
axs[1].set_title('Green Channel')

axs[2].imshow(data[:, :, 2], cmap='Blues')
axs[2].set_title('Blue Channel')

for ax in axs:
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
""" 2 e)"""
grayscale = data.mean(axis=2).astype(np.uint8) #mean axis=2 averages the red, green and blue vals for each pixel -> grayscale

plt.imshow(grayscale, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
""" 2 f)"""
binary = np.where(grayscale < 128, 0, 255).astype(np.uint8) #if pixel less than 128 -> 0 (black), otherwise 255 (white)

plt.imshow(binary, cmap='gray')
plt.title('Binary Image')
plt.axis('off')
plt.show()
```

```python
""" 2 e)"""

cropped = data[100:550, 100:700, :]

plt.imshow(cropped)
plt.title('Cropped Image')
plt.axis('off')
plt.show()
```
