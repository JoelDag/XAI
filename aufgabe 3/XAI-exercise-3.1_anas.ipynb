**e. Learn rules via the `GreedyRuleListClassifier` of `imodels`. What is the training accuracy and the test accuracy of the classifier? How can the rules be understood? For example, what does “risk” mean? What does “pts” mean? How are “risk” and “pts” related to the “coverage” and “accuracy” of a rule as introduced in the lecture?**

```python
x = df.drop("is_cnt_gt_4551", axis=1)
y = df["is_cnt_gt_4551"]

le = LabelEncoder()

x_encoded = pd.get_dummies(x)
y_encoded = le.fit_transform(y)
```

```python
x_encoded.head()
```

```python
y_encoded
```

```python
model = imodels.GreedyRuleListClassifier()
model.fit(x_encoded, y_encoded)
```

```python
print(model)
```

Risk represents how many of the positive examples the rule applies and was true.
Therefore the higher the risk the better the rule at classifying.

`pts` are the data points where the rule applies (where the condition is satisfied). For example, `cnt_2d_bfr_range_gt_4550` would include 364 pts (i.e. rows) where this condition is true.
